Download and set up data

```python
dataset = foz.load_zoo_dataset("coco-2017", max_samples=10000)
```

```python
# results = fob.compute_visualization(dataset, method = "tsne")

# ### Color by last tag
# plot = results.visualize(labels=F("tags")[-1])
# plot.show()
```

```python
# Puts 80 classes of dataset in alphabetical order
classes = dataset.distinct("ground_truth.detections.label")
print(classes)
```

```python
four.random_split(dataset, {"train": 0.8, "val": 0.2})

export_dir = "/content/gdrive/MyDrive/Colab Notebooks/51328-Proj3/coco_dataset"

for split in ["train", "val"]:
    split_view = dataset.match_tags(split)
    split_view.export(
        export_dir=export_dir,
        dataset_type=fo.types.YOLOv5Dataset,
        split=split,
        classes=classes
    )
```
